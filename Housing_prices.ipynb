```python
import numpy as np
import matplotlib.pyplot as plt 

import pandas as pd  
import seaborn as sns 

%matplotlib inline
```

```python
df_train=pd.read_csv('/content/drive/MyDrive/train.csv')
df_test=pd.read_csv('/content/drive/MyDrive/test.csv')
```

```python
print("Train data shape",df_train.shape)
print("Test data shape",df_test.shape)
```

```python
df_train_cont=df_train.select_dtypes("number")
df_test_cont=df_test.select_dtypes("number")
```

```python
df_test_clean=df_test.dropna()
```

```python
df_train1=df_train_cont.dropna()
df_test1=df_test_cont.dropna()
```

```python
print("Train data shape after cleaning",df_train1.shape)
print("Test data shape after cleaning",df_test1.shape)
```

```python
corr=df_train1.corr(method='pearson')
sns.set(rc = {'figure.figsize':(15,8)})
sns.heatmap(data=corr,annot=True)
```

```python

columns=[]
i=0
for val in corr['SalePrice']:
  if val>0.5 or val<-0.5:
    columns.append(corr['SalePrice'].index[i])
  i+=1
```

```python
columns
```

```python
import seaborn as sns
sns.pairplot(df_train1[columns])
```

```python
corr1=df_train1[columns].corr(method='pearson')
sns.set(rc = {'figure.figsize':(15,8)})
sns.heatmap(data=corr1,annot=True)
```

```python
X_train=df_train1[columns].drop('SalePrice',axis=1)
Y_train=df_train1['SalePrice']
```

```python
X_train.shape
```

```python
Y_train.describe()
```

```python
plt.hist(Y_train)
```

```python
print(max(Y_train), min(Y_train), sum(Y_train)/len(Y_train))
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_train, Y_train, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
pipeline=Pipeline([
    ('std_scalar',StandardScaler())
])
x_train=pipeline.fit_transform(x_train)
x_test=pipeline.transform(x_test)
model= LinearRegression()
model.fit(x_train,y_train)
y_train_predict = model.predict(x_train)
rmse_train = (np.sqrt(mean_squared_error(y_train, y_train_predict)))
r2_train = r2_score(y_train, y_train_predict)
print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse_train))
print('R2 score is {}'.format(r2_train))
print("\n")

y_test_predict = model.predict(x_test)
rmse_test = (np.sqrt(mean_squared_error(y_test, y_test_predict)))
r2_test = r2_score(y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse_test))
print('R2 score is {}'.format(r2_test))
```
